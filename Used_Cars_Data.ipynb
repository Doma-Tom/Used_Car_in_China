#### PLACING ALL GIVEN LOCATIONS INTO NORTH,SOUTH, EAST, AND WEST OF INDIA

```python
def Location_func(tom):
    if tom["Location"] == "Pune" or tom["Location"] == "Mumbai" or tom["Location"] == "Ahmedabad":
        return "West"
    elif tom["Location"] == "Chennai" or tom["Location"] == "Kolkata":
        return "East"
    elif tom["Location"] == "Delhi" or tom["Location"] == "Jaipur":
        return "North"
    elif tom["Location"] == "Coimbatore" or tom["Location"] == "Bangalore" or tom["Location"] =="Hyderabad" or tom["Location"] =="Kochi":
        return "South"
tom["Zones"] =tom.apply(Location_func, axis =1)
```
